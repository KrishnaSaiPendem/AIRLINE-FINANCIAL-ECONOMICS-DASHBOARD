```python
import boto3
import os

# Initialize a session using your AWS credentials
s3 = boto3.client('s3')

# Specify your S3 bucket name
bucket_name = 'cleaned-datsets-airline-project'  # Replace with your bucket name

# Specify the directory containing the CSV files
directory = "/home/sagemaker-user/Cleaned datasets_form41"

# Iterate through the directory and upload each CSV file
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)
        try:
            # Upload the file with the correct arguments
            s3.upload_file(Filename=file_path, Bucket=bucket_name, Key=filename)
            print(f"Successfully uploaded {filename} to {bucket_name}")
        except Exception as e:
            print(f"Error uploading {filename}: {e}")
```
